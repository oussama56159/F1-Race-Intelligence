# Étape 1 : Business Understanding (Compréhension des Objectifs)

Ce notebook documente la première étape de la méthodologie **CRISP-DM** appliquée à notre projet de prédiction des résultats de Formule 1.

---

## 1. Contexte du Projet

### Qu'est-ce que la Formule 1 ?
La Formule 1 (F1) est la catégorie reine du sport automobile mondial. Elle oppose des constructeurs (écuries) et des pilotes sur des circuits internationaux. Chaque week-end de course se compose de séances d'essais, de qualifications (déterminant la grille de départ) et de la course finale.

### Pourquoi prédire les résultats des courses ?
La prédiction des résultats en F1 présente plusieurs intérêts majeurs :
- **Stratégie d'écurie** : Optimiser les décisions en fonction des probabilités de succès.
- **Analyse de performance** : Identifier les facteurs clés qui mènent au podium.
- **Engagement des fans** : Améliorer l'expérience des spectateurs via des insights prédictifs.
- **Marchés de paris** : Évaluer les cotes et les probabilités objectives.

---

## 2. Objectifs Métier et Traduction Data Science

Le projet se divise en 5 objectifs spécifiques :

1. **OBJ1 : Classification Binaire (Podium)**
   - *Objectif métier* : Savoir si un pilote finira dans le top 3.
   - *Traduction* : Prédire la variable `podium` (0 ou 1).

2. **OBJ2 : Classification Multiclasse (Position Class)**
   - *Objectif métier* : Catégoriser la performance finale du pilote.
   - *Traduction* : Prédire `position_class` (Winner, Podium, Points, Outside Points).

3. **OBJ3 : Régression avec Encodage Pays (Penalty)**
   - *Objectif métier* : Estimer l'écart de performance (pénalité de position) en tenant compte de l'influence géographique.
   - *Traduction* : Prédire `finish_pos_penalty` en utilisant le pays encodé.

4. **OBJ4 : Clustering et PCA (Performance Pilote)**
   - *Objectif métier* : Segmenter les pilotes selon leur profil de performance historique.
   - *Traduction* : Appliquer K-Means/DBSCAN après une réduction de dimension (PCA).

5. **OBJ5 : Régression sans Encodage Pays (Penalty)**
   - *Objectif métier* : Estimer la pénalité de position basée uniquement sur les caractéristiques techniques et de carrière.
   - *Traduction* : Prédire `finish_pos_penalty` sans la variable pays.

## 3. Métriques de Succès

Pour évaluer la pertinence de nos modèles, nous utiliserons les métriques suivantes :

| Objectif | Type de Tâche | Métriques Clés |
| :--- | :--- | :--- |
| **OBJ1** | Classification Binaire | **F1-score**, **ROC-AUC** (gestion du déséquilibre des classes) |

## 4. Contraintes et Prévention de la Fuite de Données (Leakage)

### Règle d'or
Aucune donnée connue uniquement *après* la course ne doit être utilisée pour l'entraînement des modèles.

### Variables exclues (Leakage)
Les colonnes suivantes sont strictement interdites en tant que features :
- `positionOrder`
- `points`
- `laps`
- `rank`
- `statusId`

### Features autorisées (Pre-race)
Seules les données disponibles avant l'extinction des feux sont valides :
- Historique (year, round)
- Qualification (grid, qualifying_position)
- État du championnat (driver/constructor standing, points, wins)
- Environnement (country - selon OBJ)

## 5. Cycle de Vie CRISP-DM

Le projet suit le cycle itératif **CRISP-DM** (*Cross-Industry Standard Process for Data Mining*) :

1. **Business Understanding** (Ce notebook) : Définition des objectifs et des succès.
2. **Data Understanding** : Exploration du dataset F1 (26,499 lignes) et analyse descriptive.
3. **Data Preparation** : Nettoyage, encodage des variables catégorielles et feature engineering.
4. **Modeling** : Entraînement des algorithmes (Classification, Régression, Clustering).
5. **Evaluation** : Comparaison des performances par rapport aux métriques définies.
6. **Deployment** : Mise à disposition des modèles ou des conclusions stratégiques.
